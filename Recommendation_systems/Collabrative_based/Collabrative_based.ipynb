# Collabrative\-based Filtering Movie\_Recommendation

```
# load libraries

import pandas as pd
```

```
ratings = pd.read_csv('ratings.csv')
```

```
ratings.head()
```

## Transforming the data for the "Surprise"

```
ratings = ratings[['userId', 'movieId', 'rating']]
```

```
ratings.head()
```

```
!pip install scikit-surprise
```

```
# Load the dataset from the ratings to the surprise format.

from surprise import Dataset, Reader

reader = Reader(rating_scale=(1, 5))
dataset = Dataset.load_from_df(ratings, reader)
dataset
```

## Build training set

```
trainset = dataset.build_full_trainset()
```

```
# Model definition

from surprise import SVD

svd = SVD()
```

```
svd.fit(trainset)
```

```
svd.predict(1, 1061)
```

## Validation

```
actual_data = pd.read_csv('ratings.csv')
```

```
from surprise import model_selection

cross_validation = model_selection.cross_validate(svd, data=dataset, measures=['rmse'])
```
